## Lets try to understand which edges to add to lower or increase the hyperbolicity

```python

import random

def gromov_hyperbolicity(G):
    dist = compute_distance_nodes(G)
    delta_max, delta_mean = compute_gromov_hyperbolicity(dist)
    return delta_max

def test_gromov_edge_addition(G, increase=True):
    original_gromov = gromov_hyperbolicity(G)
    nodes = list(G.nodes)
    candidate_edges = [(u, v) for i, u in enumerate(nodes) for v in nodes[i+1:] if not G.has_edge(u, v)]
    
    result_edges = []
    
    for u, v in candidate_edges:
        G.add_edge(u, v)
        new_gromov = gromov_hyperbolicity(G)
        if (increase and new_gromov > original_gromov) or (not increase and new_gromov < original_gromov):
            result_edges.append((u, v))
        G.remove_edge(u, v)
    
    # Se non ci sono archi che soddisfano la condizione, aggiungi/rimuovi uno a caso
    if not result_edges:
        if increase and candidate_edges:
            # aggiungi un arco a caso
            edge_to_add = random.choice(candidate_edges)
            G.add_edge(*edge_to_add)
            print(f"Aggiunto arco a caso: {edge_to_add}")
            result_edges.append(edge_to_add)
        elif not increase and G.number_of_edges() > 0:
            # rimuovi un arco a caso
            edge_to_remove = random.choice(list(G.edges))
            G.remove_edge(*edge_to_remove)
            print(f"Rimosso arco a caso: {edge_to_remove}")
            result_edges.append(edge_to_remove)
    
    return result_edges

# Esempio d'uso
G = nx.cycle_graph(5)  # grafo di esempio
edges_to_add_or_remove = test_gromov_edge_addition(G, increase=True)
print("Archi selezionati:", edges_to_add_or_remove)
```
