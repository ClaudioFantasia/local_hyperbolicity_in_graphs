## Gromov Hyperbolicity: The Four-Point Condition

A metric space $(X, d)$ is said to be **$\delta$-hyperbolic** if for any four points $x, y, z, w \in X$, the following inequality holds:

$$d(x, y) + d(z, w) \le \max \{ d(x, z) + d(y, w), d(x, w) + d(y, z) \} + 2\delta$$

### Understanding the Triple Sums

We define the three possible pairings of the four points into two pairs:

1. $S_1 = d(x, y) + d(z, w)$
2. $S_2 = d(x, z) + d(y, w)$
3. $S_3 = d(x, w) + d(y, z)$

$S_1, S_2, S_3$ sorted in non-increasing order such that:
$$S_1 \ge S_2 \ge S_3$$

The four-point condition explain how close these two distances are close:
$$S_1 - S_2 \le 2\delta$$

---

You can compute the gromov hyperbolicity computing gromov product over every possible permutations and taking the max or you can computing it over every possible combinations and taking the two largest sum because of symmetries.

## Lets try to understand which edges to add to lower or increase the hyperbolicity

# this tests are on another notebook

```python

import random


def test_gromov_edge_addition(G, increase=True, original_gromov=0):
    nodes = list(G.nodes)
    candidate_edges = [(u, v) for i, u in enumerate(nodes) for v in nodes[i+1:] if not G.has_edge(u, v)]
    
    result_edges = []
    
    for u, v in candidate_edges:
        G.add_edge(u, v)
        new_gromov = gromov_hyperbolicity(G)
        if (increase and new_gromov > original_gromov) or (not increase and new_gromov < original_gromov):
            result_edges.append((u, v))
        G.remove_edge(u, v)

    if not result_edges:
        if increase and candidate_edges:            
            edge_to_add = random.choice(candidate_edges)
            G.add_edge(*edge_to_add)
            print(f"Added a random edge from candidates: {edge_to_add}")
            #result_edges.append(edge_to_add)
        elif not increase and G.number_of_edges() > 0:
            edge_to_remove = random.choice(list(G.edges))
            G.remove_edge(*edge_to_remove)
            print(f"Removed a random edge: {edge_to_remove}")
            #result_edges.append(edge_to_remove)
    
    return result_edges


```

```python
n = 10
G = create_cycle_graph(n=n)
pos = draw_layout(G, pos=None)
draw_graph(G, pos)

print(f"Original gromov hyperbolicity: {gromov_hyperbolicity(G)}")
```

## here we see that we cannot increase the value of initial gromov hyperbolicity of 2.0

```python
result_edges = []
gromov_measure = gromov_hyperbolicity(G)


while (not result_edges) and len(G.edges) != (n * (n - 1)) // 2:
    result_edges = test_gromov_edge_addition(G, increase=True, original_gromov=gromov_measure)
```

```python
print(f"Edges that increase gromov hyperbolicity: {result_edges}")
if result_edges:
    G.add_edge(*result_edges[0])
    print('added the first edge in the candidates list')
    draw_graph(G, pos)
    print(f"New gromov hyperbolicity: {gromov_hyperbolicity(G)}")
```

```python
def tmp_draw_graph(
    G,
    pos,
    ax=None,
    highlight_edges=None,
    with_labels=True,
    node_color="lightblue",
    edge_color="gray",
    highlight_color="red",
    node_size=500,
    font_size=10,
    title=None
):
    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 6))

    # Nodi
    nx.draw_networkx_nodes(G, pos, node_color=node_color, node_size=node_size, ax=ax)

    # Archi normali
    if highlight_edges:
        normal_edges = [e for e in G.edges if e not in highlight_edges and (e[1], e[0]) not in highlight_edges]
    else:
        normal_edges = list(G.edges)

    nx.draw_networkx_edges(G, pos, edgelist=normal_edges, edge_color=edge_color, ax=ax)

    # Archi evidenziati
    if highlight_edges:
        nx.draw_networkx_edges(
            G,
            pos,
            edgelist=highlight_edges,
            edge_color=highlight_color,
            width=2.5,
            ax=ax
        )

    # Labels
    if with_labels:
        nx.draw_networkx_labels(G, pos, font_size=font_size, ax=ax)

    if title:
        ax.set_title(title)

    ax.axis("off")
    return ax
```
